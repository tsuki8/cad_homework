# 构造三阶B样条节点向量，写出各阶上的基函数

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import BSpline

# 定义节点向量
t = np.array([0, 1, 2, 3, 4, 5, 6, 7])

# 定义B样条对象
k = 3  # 三阶B样条
b_spline = BSpline(t, np.eye(len(t)-k+1), k)

# 生成绘图用的x值
x = np.linspace(0, 6, 100)

# 计算每个基函数的值
y = b_spline(x)

# 绘制基函数
plt.figure(figsize=(10, 8))
for i in range(len(t)-k+1):
    plt.plot(x, y[:, i], label=f'B_{i,3}(t)')

plt.title('Cubic B-Spline Basis Functions')
plt.xlabel('t')
plt.ylabel('B-spline basis function value')
plt.legend()
plt.grid(True)
plt.show()
```

# 画三阶具有7个控制点的b样条曲线, 分别有节点重一次，控制点重一次，求导一次

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import BSpline
```

## 节点重一次

```python
# 7个控制点
control_points = np.array([[0, 0], [1, 2], [2, 0], [3, 3], [4, 3], [4, 1], [6, 0]])

# 节点矢量 (knot vector)，某个节点重合一次
knots = np.array([0,0,0,0,1,2,2,4,4,4,4])

# 生成B样条，k=3表示三阶
# spl_x = BSpline(knots, control_points[:,0], k=3)
# spl_y = BSpline(knots, control_points[:,1], k=3)
spl = BSpline(knots, control_points, k=3)

# t取值范围（参数化范围）
t = np.linspace(0, 4, 100)

# 绘制B样条曲线
plt.figure(figsize=(8, 6))
plt.plot(*spl(t).T, label="B-Spline Curve with Repeated Knot")
plt.scatter(control_points[:, 0], control_points[:, 1], color='red', label='Control Points')
# 绘制控制点之间的虚线
plt.plot([0,2],'--', color='gray', alpha=0.5)
plt.plot([1,3],[2,3],'--', color='gray', alpha=0.5)
plt.plot([3,4],[3,3],'--', color='gray', alpha=0.5)
plt.plot([4,6],[3,0],'--', color='gray', alpha=0.5)
plt.title('B-Spline Curve (Repeated Knot)')
plt.legend()
plt.grid(True)
plt.savefig('HW4_1_1.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

# 控制点重一次的三阶B样条曲线

```python
# 7个控制点，其中一个控制点重复
control_points = np.array([[0, 0], [1, 2], [2, 0], [3, 3], [3, 3], [4, 1], [6, 0]])

# 节点矢量 (knot vector)
knots = np.array([0,0,0,0,1,2,3,4,4,4,4])

# 生成B样条，k=3表示三阶
spl = BSpline(knots, control_points, k=3)

# t取值范围
t = np.linspace(0, 4, 100)

# 绘制B样条曲线
plt.figure(figsize=(8, 6))
plt.plot(*spl(t).T, label="B-Spline Curve with Repeated Control Point")
plt.scatter(control_points[:, 0], control_points[:, 1], color='red', label='Control Points')
plt.plot([0,2],'--', color='gray', alpha=0.5)
plt.plot([1,3],[2,3],'--', color='gray', alpha=0.5)
plt.plot([3,6],[3,0],'--', color='gray', alpha=0.5)
plt.title('B-Spline Curve (Repeated Control Point)')
plt.legend()
plt.grid(True)
plt.savefig('HW4_1_2.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

# 求导一次的B样条曲线

```python
# 7个控制点
control_points = np.array([[0, 0], [1, 2], [2, 0], [3, 3], [4, 3], [4, 1], [6, 0]])

# 节点矢量 (knot vector)
knots = np.array([0,0,0,0,1,2,3,4,4,4,4])

# 生成B样条，k=3表示三阶
spl = BSpline(knots, control_points, k=3)

# 对B样条曲线求导
spl_derivative = spl.derivative()

# t取值范围
t = np.linspace(0, 4, 100)

# 绘制原始B样条曲线
plt.figure(figsize=(8, 6))
plt.plot(*spl(t).T, label="Original B-Spline Curve")
plt.plot(*spl_derivative(t).T, label="First Derivative of B-Spline", linestyle='--')
plt.scatter(control_points[:, 0], control_points[:, 1], color='red', label='Control Points')
plt.plot([0,2],'--', color='gray', alpha=0.5)
plt.plot([1,3],[2,3],'--', color='gray', alpha=0.5)
plt.plot([3,4],[3,3],'--', color='gray', alpha=0.5)
plt.plot([4,6],[3,0],'--', color='gray', alpha=0.5)
plt.title('B-Spline Curve and Its Derivative')
plt.legend()
plt.grid(True)
plt.savefig('HW4_1_3.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
